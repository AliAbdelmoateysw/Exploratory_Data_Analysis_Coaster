```python
df= df[['coaster_name', #'Length', 'Speed', 
        'Location', 'Status',
    #    'Opening date','Type', 
        'Manufacturer',
    #     'Height restriction', 'Model', 'Height',
    #    'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #    'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #    'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #    'Track layout', 'Fastrack available', 'Soft opening date.1',
    #    'Closing date', 'Opened', 'Replaced by', 'Website',
    #    'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #    'Single rider line available', 'Restraint Style',
    #    'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',
    #    'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
    #    'height_value', 'height_unit', 
        'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.dtypes
```

```python
df['opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
```

```python
df.columns
```

```python
#Rename Columns
df = df.rename(columns={'coaster_name':'Coaster_Name' , 'year_introduced':'Year_Introduced',
      'opening_date_clean':'Opening_Date', 'speed_mph':'Speed_mph',
       'height_ft':'Height_ft', 'Inversions_clean':'Inversions', 'Gforce_clean':'Gforce'})
```

```python
df.isna().sum()
```

```python
#Check for Duplicate coaster name
df.loc[df.duplicated(subset='Coaster_Name')] .head(5)
```

```python
df.query("Coaster_Name=='Crystal Beach Cyclone'")
```

```python
df =  df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True)
```

Feature Relationships

```python
df.plot(kind='scatter',x='Speed_mph',y='Height_ft',title='Coaster Speed (mph) vs Height')
plt.show()
```

```python
ax= sns.scatterplot(x='Speed_mph',y='Height_ft',data=df,hue='Year_Introduced')
ax.set_title('Coaster Speed (mph) vs Height')
plt.show()
```

```python
sns.pairplot(df,vars=['Year_Introduced','Speed_mph','Height_ft','Inversions','Gforce'],hue='Type_Main')
plt.show()
```

```python
df_corr = df[['Year_Introduced','Speed_mph','Height_ft','Inversions','Gforce']].dropna().corr()
```

```python
sns.heatmap(df_corr,annot=True,cmap='YlGnBu')
plt.show()
```

Ask Questions about data

```python
df['Location'].value_counts()
```

```python
#location of fastest roller coaster
ax =df.query('Location !="Other"')\
    .groupby('Location')['Speed_mph']\
    .agg(['mean','count']).query("count >=10")\
    .sort_values('mean').plot(kind='barh',figsize=(20,7),title='Average Coast Speed by Location')
ax.set_xlabel('Average Coast Speed by Location')
plt.show()
```
